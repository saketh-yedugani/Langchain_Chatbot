```python
import os
from dotenv import load_dotenv
load_dotenv() ## aloading all the environment variable

groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(model="llama-3.3-70b-versatile",groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage
model.invoke([HumanMessage(content="Hi , My name is Saketh and I am a Software Engineer")])
```

```python
from langchain_core.messages import AIMessage
model.invoke(
    [
        HumanMessage(content="Hi , My name is Saketh and I am a Software Engineer"),
        AIMessage(content="Hello Saketh! It's nice to meet you. \n\nAs a Software Engineer, what kind of projects are you working on these days? \n\nI'm always eager to learn more about the exciting work being done in the field of AI.\n"),
        HumanMessage(content="Hey What's my name and what do I do?")
    ]
)
```

```python
##Here we create a function which verify's if the session ID is in the stor if not it will add a new key,value pair in the dictionary

from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store={}

def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id]=ChatMessageHistory()
    return store[session_id]

with_message_history=RunnableWithMessageHistory(model,get_session_history)
```

```python
config={"configurable":{"session_id":"chat1"}}
```

```python
response=with_message_history.invoke(
    [HumanMessage(content="Hi , My name is Saketh and I am a Software Engineer")],
    config=config
)
```

```python
with_message_history.invoke(
    [HumanMessage(content="What's my name?")],
    config=config,
)
```

```python
## change the config-->session id
##Since we changed the session id the model does'nt remember our earlier chats
config1={"configurable":{"session_id":"chat2"}}
response=with_message_history.invoke(
    [HumanMessage(content="Whats my name")],
    config=config1
)
response.content
```

```python
## Here we set up a place holder which creates called message where in the message we give in goes into thet place holder.
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant.Amnswer all the question to the nest of your ability"),
        MessagesPlaceholder(variable_name="messages")
    ]
)

chain=prompt|model
```

```python
#Here the message goes into the prompt place holder
chain.invoke({"messages":[HumanMessage(content="Hi My name is Saketh")]})
```

```python
with_message_history=RunnableWithMessageHistory(chain,get_session_history)
```

```python
config = {"configurable": {"session_id": "chat3"}}
response=with_message_history.invoke(
    [HumanMessage(content="Hi My name is Saketh")],
    config=config
)

response
```

```python
response = with_message_history.invoke(
    [HumanMessage(content="What's my name?")],
    config=config,
)

response.content
```

```python
## In this example we add 2 place holders which is message and language place holders

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a helpful assistant. Answer all questions to the best of your ability in {language}.",
        ),
        MessagesPlaceholder(variable_name="messages"),
    ]
)

chain = prompt | model
```

```python
response=chain.invoke({"messages":[HumanMessage(content="Hi My name is Saketh")],"language":"Hindi"})
response.content
```

```python
## We are adding a set of message history and also add a trimmer which only considers the message in that limit of the words
from langchain_core.messages import SystemMessage,trim_messages
trimmer=trim_messages(
    max_tokens=45,
    strategy="last",
    token_counter=model,
    include_system=True,
    allow_partial=False,
    start_on="human"
)
messages = [
    SystemMessage(content="you're a good assistant"),
    HumanMessage(content="hi! I'm bob"),
    AIMessage(content="hi!"),
    HumanMessage(content="I like vanilla ice cream"),
    AIMessage(content="nice"),
    HumanMessage(content="whats 2 + 2"),
    AIMessage(content="4"),
    HumanMessage(content="thanks"),
    AIMessage(content="no problem!"),
    HumanMessage(content="having fun?"),
    AIMessage(content="yes!"),
]
trimmer.invoke(messages)
```

```python
#We created a chain which considers these messages and them go through a trimmer and then a promptand the model
from operator import itemgetter

from langchain_core.runnables import RunnablePassthrough

chain=(
    RunnablePassthrough.assign(messages=itemgetter("messages")|trimmer)
    | prompt
    | model
    
)

response=chain.invoke(
    {
    "messages":messages + [HumanMessage(content="What ice cream do i like")],
    "language":"English"
    }
)
response.content
```

```python
response = chain.invoke(
    {
        "messages": messages + [HumanMessage(content="what math problem did i ask")],
        "language": "English",
    }
)
response.content
```

```python
## Lets wrap this in the Message History
with_message_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages",
)
config={"configurable":{"session_id":"chat5"}}
```

```python
response = with_message_history.invoke(
    {
        "messages": messages + [HumanMessage(content="whats my name?")],
        "language": "English",
    },
    config=config,
)

response.content
```
